```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
# we merged zomato dataset with Country-Code dataset. so that we can get countries along with data in zomato.
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(3)
```

```python
#check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
```

```python
# pie chart - Top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

**Observation:** *Zomato maximum records or transactions are from India then USA and then United Kindom*

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# **Observation**

### 1. When Rating is between 4.5-4.9 ---> *Excellent*
### 2. When Rating is between 4.0-4.4 ---> *Very Good*
### 3. When Rating is between 3.5-3.9 ---> *Good*
### 4. When Rating is between 2.5-3.4 ---> *Average*
### 5. When Rating is between 1.8-2.4 ---> *Poor*
### 6. When Rating is 0.0 ---> *Not Rated*

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['grey','Red','Orange','Yellow','lightgreen','Green'])
```

# Not Rated count is very High
# Maximum number of rating are between 2.5 to 3.4

```python
#Count Plot
sns.countplot(x="Rating color",data=ratings,palette=['grey','Red','Orange','Yellow','lightgreen','Green'])
```

```python
#find 0 rated countries
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:"Country Count"})
```

# Maximum no of 0 ratings are from INDIA

```python
# currency and country relation
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index().rename(columns={0:"Currency Count"})
```

```python
## Which Countries do have online deliveries option
final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```
